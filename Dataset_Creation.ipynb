```python
import scipy.io
import numpy as np
import pandas as pd
```

```python
mat1 = scipy.io.loadmat('Data/data-science-P1.mat')
```

```python
dat = []
data = mat1['data']
info = mat1['info']
meta = mat1['meta']
coord_to_col = meta[0][0][-1]
for i in range(data.shape[0]):
    print(i+1)
    temp = []
    first = data[i, :]
    temp.append(first[0][0])
    val = info[0][i][2][0]
    object_type = info[0][i][0][0]
    
    brain3d = list(np.zeros((meta[0][0][4][0][0], meta[0][0][5][0][0], meta[0][0][6][0][0])))
    for i in range(coord_to_col.shape[0]):
        for j in range(coord_to_col.shape[1]):
            for k in range(coord_to_col.shape[2]):
                vox = coord_to_col[i][j][k]
                if vox == 0:
                    brain3d[i][j][k] = 0
                brain3d[i][j][k] = first[0][0][vox-1]
    temp.append(brain3d)
    temp.append(object_type)
    temp.append(val)
    dat.append(temp)
df = pd.DataFrame(dat, columns=['FMRI_Data', '3D_FMRI_Data', 'Object_Type', 'Object'])
df
```

```python
df.to_csv('Data/data_csv/P1.csv')
```
